```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image=cv2.imread('./images/obama.jpg',0)
print("1.ORIGINAL \n2.THRESHOLD \n3.ADAPTIVE THRESHOLD\n4.GASSIAN_BLUR \n4.ADAPTIVE THRESHOLD \
      \n5.GRAYSCALED\n6.CANNY\n7.LAPLACIAN\n8.SOBEL(SOBEL_X<SOBEL_Y AND SOBEL_OR)\n9.ROTATION\n10.SMOOTHING")
'''def original():
    ##for original
    image=cv2.imread('./images/obama.jpg')## making and object
    cv2.imshow('ORIGINAL',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()'''
def original():
    plt.subplot(2,2,1),plt.imshow(image,cmap = 'gray')
    plt.title('Original'), plt.xticks([]), plt.yticks([])
    plt.show()
    

    
   
def simple_thres():
    img = cv2.imread('./images/obama.jpg')
    ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
    ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
    ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
    ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

    titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
    images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

    for i in range(6):
        plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
        plt.title(titles[i])
        plt.xticks([]),plt.yticks([])

    plt.show()
def adap_thres():
    img = cv2.imread('./images/obama.jpg',0)
    img = cv2.medianBlur(img,5)

    ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
                cv2.THRESH_BINARY,11,2)
    th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
                cv2.THRESH_BINARY,11,2)

    titles = ['Original Image', 'Global Thresholding (v = 127)',
                'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
    images = [img, th1, th2, th3]

    for i in range(4):
        plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
        plt.title(titles[i])
        plt.xticks([]),plt.yticks([])
    plt.show()




def gassian_blur():
    image1=cv2.GaussianBlur(image,(3,3),0)
    cv2.imshow('Gaussian blur',image1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    

    
    
def grayscale():
    image=cv2.imread('./images/obama.jpg')## making and object
    gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    cv2.imshow('GrayScale',gray_image)
    cv2.waitKey()## A trigger to open the image.
    cv2.destroyAllWindows()
    
    
def canny():
    #img = cv2.imread('./images/obama.jpg',0)
    edges = cv2.Canny(image,100,200)

    plt.subplot(121),plt.imshow(image,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

    plt.show()
    
def laplacian():
    laplacian = cv2.Laplacian(image,cv2.CV_64F)
    plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
    plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
    plt.show()
    
    
def sobel():
        
    def sobel_x():
        sobelx = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=5)

        plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
        plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
        plt.show()

    def sobel_y():
        sobely = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=5)

        plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
        plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
        plt.show()
    print("1.Sobel X\n2.Sobel Y")
    x=int(input("ENTER NUMBER OF SOBEL FILTER YOU WANT TO APPLY"))
    for i in range(0,x):
        
        n = int(input("CHOOSE THE SOBEL FILTER  :  "))
        if n == 1:
            sobel_x()
        elif n== 2:
            sobel_y()
        


    
def rotation():
    print("1.LEFT \n2.RIGHT\n3.UPSIDE ")
    def rotate_left():
        rows,cols = image.shape
        M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
        dst = cv2.warpAffine(image,M,(cols,rows))

        cv2.imshow('image',dst)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    def rotate_right():
        rows,cols = image.shape
        M = cv2.getRotationMatrix2D((cols/2,rows/2),-90,1)
        dst = cv2.warpAffine(image,M,(cols,rows))

        cv2.imshow('image',dst)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    def rotate_upside():
        rows,cols = image.shape
        M = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
        dst = cv2.warpAffine(image,M,(cols,rows))

        cv2.imshow('image',dst)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    x = int(input("ENTER THE NUMBER OF DIRECTIONS :   "))
    for i in range(0,x):
        n = int(input("CHOOSE THE DIRECTION :  "))
        if n == 1:
            rotate_left()
        elif n== 2:
            rotate_right()
        elif n == 3:
            rotate_upside()
def smoothing():
    def averaging():
        img = cv2.imread('./images/obama.jpg')

        blur3 = cv2.blur(img,(5,5))

        plt.subplot(121),plt.imshow(img),plt.title('Original')
        plt.xticks([]), plt.yticks([])
        plt.subplot(122),plt.imshow(blur3),plt.title('Blurred')
        plt.xticks([]), plt.yticks([])
        plt.show()
    def gaussian_filtering():
        img = cv2.imread('./images/obama.jpg')

        blur2 = cv2.GaussianBlur(img,(5,5),0)

        plt.subplot(121),plt.imshow(img),plt.title('Original')
        plt.xticks([]), plt.yticks([])
        plt.subplot(122),plt.imshow(blur2),plt.title('Blurred')
        plt.xticks([]), plt.yticks([])
        plt.show()
    def Median_Filtering():
        img = cv2.imread('./images/obama.jpg')
        median = cv2.medianBlur(img,5)
        plt.subplot(121),plt.imshow(img),plt.title('Original')
        plt.xticks([]), plt.yticks([])
        plt.subplot(122),plt.imshow(median),plt.title('Median')
        plt.xticks([]), plt.yticks([])
        plt.show()
    def Bilateral_Filtering():
        img = cv2.imread('./images/obama.jpg')
        blur1 = cv2.bilateralFilter(img,9,75,75)
        plt.subplot(121),plt.imshow(img),plt.title('Original')
        plt.xticks([]), plt.yticks([])
        plt.subplot(122),plt.imshow(blur1),plt.title('Blurred')
        plt.xticks([]), plt.yticks([])
        plt.show()
        
        
    print("1.AVERAGING\n2.GAUSSIAN FILTERING\n3.MEDIAN FILTERING\n4.BILATERAL FILTERING")
    x=int(input("ENTER NUMBER OF SMOOTHING FILTER YOU WANT TO APPLY"))
    for i in range(0,x):
        n = int(input("CHOOSE THE SMOOTHING FILTER  :  "))
        if n == 1:
            averaging()
        elif n == 2:
            gaussian_filtering()
        elif n == 3:
            Median_Filtering()
        elif n== 4:
            Bilateral_Filtering()

    
    
x = int(input("ENTER THE NUMBER OF FILTERS YOU WANT TO APPLY :   "))
for i in range(0,x):
    n = int(input("CHOOSE THE FILTER  :  "))
    if n == 1:
        original()
    elif n== 2:
        simple_thres()
    elif n == 3:
        adap_thres()
    elif n == 4:
        gassian_blur()

    elif n == 5:
        grayscale()
    elif n == 6:
        canny()
    elif n == 7:
        laplacian()
    elif n == 8:
        sobel()
    elif n == 9:
        rotation()
    elif n==10:
        smoothing()
       
```
